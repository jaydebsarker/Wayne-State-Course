```python
import numpy as np
from numpy import *
from matplotlib import pyplot as plt
```

```python
X = []
y = []
```

```python
def loadDataSet():
    f=open('Question5.txt')
    # Read data line by line and use strip to remove the Spaces 
    for line in f.readlines():
        nline=line.strip().split()
        # X has two columns
        X.append([float(nline[0]),float(nline[1])])
        y.append(int(nline[2]))
    return mat(X).T,mat(y)
```

```python
X,y = loadDataSet()
```

```python
# Define sigmoid function
def sigmoid(x):
    return 1.0/(1+exp(-x))
```

## gradient descent

```python
# Define Logistic Regression function
def Logistic(X,y,W,b,n,alpha,iterations):
    J = zeros((iterations,1))
    for i in range(iterations):   
        # step1 forward propagation
        y_hat = sigmoid(np.dot(W,X)+b)
        # compute cost function
        J[i]=-1/n *(np.dot(y,np.log(y_hat.T))+np.dot((1-y),np.log((1-y_hat).T)))
        # step2 back propagation
        dy = y_hat-y
        dW = 1/n * np.dot(dy,X.T)
        db = 1/n * np.sum(dy)
        # step3 gradient descent
        W = W-alpha*dW
        b = b-alpha*db
    return y_hat,W,b,J
```

## Plotting the decision boundary# 

```python
# Plot figures
def plotBestFit(X,y,J,W,b,n,y_hat):
    
    # Plot cost function figure
    fig1 = plt.figure(1)
    plt.plot(J)
    plt.title(u'cost function result')
    plt.xlabel(u'epoch')
    plt.ylabel(u'cost')
    
    # Plot the final classification figure
    # Divided data into two kinds of points based on the labels
    xcord1=[]; ycord1=[]
    xcord2=[]; ycord2=[]
    for i in range(n):
        if int(y[0,i])==1:
            xcord1.append(X[0,i])
            ycord1.append(X[1,i])
        else:
            xcord2.append(X[0,i])
            ycord2.append(X[1,i])
            
    fig2 = plt.figure(2)
    plt.scatter(xcord1,ycord1,c='b',marker='o')
    plt.scatter(xcord2,ycord2,c='r',marker='s')
    x=linspace(-3,3,100).reshape(100,1) 
    y=(-b-W[0,0]*x)/W[0,1]
    plt.plot(x,y,c='y')
    plt.title(u'Logistic Regression Classification Result')
    plt.xlabel(u'X')
    plt.ylabel(u'y')
    
    plt.show()
```

```python
num = X.shape[0]  # number of features
n = X.shape[1] # number of samples

# Initianlize the weights and bias
W=np.random.randn(1,num)*0.01
b=0

# Learning rate
alpha=0.05

# Iterations
iterations = 5000
```

```python
y_hat,W,b,J = Logistic(X, y, W, b, n, alpha, iterations)
```

```python
plotBestFit(X, y, J, W, b, n, y_hat)
```
